```python
import warnings
import pandas as pd
import numpy as np
warnings.filterwarnings('ignore')
from plotnine import *
```

```python
read_file = 'pickle_data/201601.xlsx.pkl'
```

```python
f1 = pd.read_pickle(read_file)
f1['入口日期及时间'] = pd.to_datetime(f1['入口日期及时间'])
f1['出口日期及时间'] = pd.to_datetime(f1['出口日期及时间'])
f1['出口日期及时间']
```

```python
f1.columns
```

```python
renamecolIndex = ['unnamed: 0 ','entry network number', 'entry station number', 'entry date and time', 'exit network number', 'exit station number',
'exit date and time', 'exit lane number', 'vehicle model code', 'vehicle type code', 'mileage', 'total axle number', 'axle type and axle weight', 'total vehicle and cargo weight', 'weight limit',
'overrun rate', 'green-traffic vehicle code', 'free type code', 'path identification', 'etc Lane code', 'etc vehicle electronic tag OBU number',
'payment method code', 'license plate number', 'driving time', 'average speed of passing lane (km / h)']
```

```python
f1.columns = renamecolIndex 
```

```python
f1["entrymonth"] = f1['entry date and time'].dt.month
f1['weekday'] = f1['entry date and time'].dt.weekday
f1['entryday'] = f1['entry date and time'].dt.day
f1['entryhour'] = f1['entry date and time'].dt.hour

f1["exitmonth"] = f1['exit date and time'].dt.month
f1['exitday'] = f1['exit date and time'].dt.day
f1['exithour'] = f1['exit date and time'].dt.hour
```

```python
f1['green-traffic vehicle code'] = f1['green-traffic vehicle code'].astype('str')
```

```python
ggplot(f1, aes(x='weekday', fill = 'green-traffic vehicle code')) + \
    geom_bar(stat = 'count')
```

```python
ggplot(f1, aes(x='entryhour', fill = 'green-traffic vehicle code')) + \
    geom_bar(stat = 'count')
```

```python
ggplot(f1, aes(x='exithour', fill = 'green-traffic vehicle code')) + \
    geom_bar(stat = 'count')
```

```python
ggplot(f1, aes(x='exithour', fill = 'green-traffic vehicle code')) + \
    geom_bar(stat = 'count')
```
