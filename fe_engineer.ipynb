```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 120)
pd.set_option('display.max_rows', 120)
```

```python
Jan = pd.read_csv("cleandata/Jan.csv")
March = pd.read_csv("cleandata/March.csv")
April = pd.read_csv("cleandata/April.csv")
May = pd.read_csv("cleandata/May.csv")

outliersJan = pd.read_csv("outliers/Jan.csv")
outliersMarch = pd.read_csv("outliers/March.csv")
outliersApril = pd.read_csv("outliers/April.csv")
outliersMay = pd.read_csv("outliers/May.csv")
```

```python
data = Jan.append(March)
data = data.append(April)
data = data.append(May)

outliersdata = outliersJan.append(outliersMarch)
outliersdata = outliersdata.append(outliersApril)
outliersdata = outliersdata.append(outliersMay)
```

```python
redundant_col = ['Unnamed: 0','入口网络编号', '出口网络编号', '出口车道编号','车种代码','免费类型代码','支付方式代码','cluster','车牌号','路径标识', 'ETC车辆电子标签OBU编号']
data = data.drop(redundant_col, axis=1)
outliersdata = outliersdata.drop(redundant_col, axis=1)
display(data.head(2))
```

Usually, vehicles carrying agricultural food won't drive for several days, so we do not need to consider the month and even week difference.

```python
from sklearn import preprocessing

data['入口站编号'] = data['入口站编号'].astype('str')
data['出口站编号'] = data['出口站编号'].astype('str')
data['轴型及轴重'] = data['轴型及轴重'].astype('str')

outliersdata['入口站编号'] = outliersdata['入口站编号'].astype('str')
outliersdata['出口站编号'] = outliersdata['出口站编号'].astype('str')
outliersdata['轴型及轴重'] = outliersdata['轴型及轴重'].astype('str')


leEntryCode = preprocessing.LabelEncoder()
data['入口站编号'] = leEntryCode.fit_transform(data['入口站编号'])
outliersdata['入口站编号'] = leEntryCode.fit_transform(outliersdata['入口站编号'])

leExitCode = preprocessing.LabelEncoder()
data['出口站编号'] = leExitCode.fit_transform(data['出口站编号'])
outliersdata['出口站编号'] = leExitCode.fit_transform(outliersdata['出口站编号'])

leVCode = preprocessing.LabelEncoder()
data['轴型及轴重'] = leVCode.fit_transform(data['轴型及轴重'])
outliersdata['轴型及轴重'] = leVCode.fit_transform(outliersdata['轴型及轴重'])
```
